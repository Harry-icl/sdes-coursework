```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
spx_data = pd.read_csv('S&P 500 Historical Data.csv', thousands=',')
```

```python
spx_data['log_returns'] = np.nan
spx_data.loc[range(len(spx_data) - 1), 'log_returns'] = np.log(spx_data.Price.values[:-1] / spx_data.Price.values[1:])
```

```python
spx_data
```

```python
periods = {
    '1 week': 5,
    '2 weeks': 10,
    '1 month': 21,
    '3 months': 63,
    '6 months': 126,
    '1 year': 252,
    '3 years': 756,
    '5 years': 1260,
    'whole': 2518
}
```

```python
results_table = pd.DataFrame(columns=['mean', 'annualised vol', 'skew', 'kurt'], index=periods.keys())
for period_name, days in periods.items():
    sample = spx_data.log_returns[:days]
    results_table.loc[period_name] = [
        sample.mean(),
        sample.std()*np.sqrt(252),
        sample.skew(),
        sample.kurtosis()
    ]
    sm.qqplot(sample)
    plt.title(f"QQplot for period of {period_name}")
    plt.show()
```
